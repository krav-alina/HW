```python
def make_translation(source, target):
    assert len(source) == len(target)
    return {
        ord(a): ord(b)
        for a, b in zip(source, target)
    }
```

```python
def preprocess(text):
    text = text.replace('\xa0', ' ')
    text = text.replace('\xad', '')
    return text
```

```python
texts = []
for path in sorted(glob('texts/*.txt')):
    with open(path, encoding = 'UTF-8') as file:
        text = file.read()
        text = preprocess(text)
        texts.append(text)
```

```python
import re

def find_person_characteristics(text):
    a = re.findall(r'(мужчина|женщина), (.+?)\n',text, re.I | re.S)
    b = re.findall(r'\d+\s\d{3}\nруб.', text, re.I | re.S)
    c = re.findall(r'(опыт работы.+?)\n', text, re.I | re.S)
    d = a+b+c
    return d
```

```python
def maybe_ingredient(line):
    match = re.search(r'\d', line)
    size = len(line) <= 50
    return match and size
```

```python
def show_matches(rule, *lines):
    parser = Parser(rule)
    for line in lines:
        matches = parser.findall(line)
        matches = sorted(matches, key=lambda _: _.span)
        spans = [_.span for _ in matches]
        show_markup(line, spans)
        if matches:
            facts = [_.fact for _ in matches]
            if len(facts) == 1:
                facts = facts[0]
            display(facts)


INT = type('INT')
NOUN = gram('NOUN')
ADJF = gram('ADJF')
PRTF = gram('PRTF')
GENT = gram('gent')
DOT = eq('.')

TOKENIZER = MorphTokenizer()
```

```python
DASH = eq('—')
```

```python
Experience = fact('Experience',['type'])

YEARS = morph_pipeline(['год','года','лет'])
NAME = morph_pipeline(['опыт','Опыт'])
NAME1 = morph_pipeline(['работа','работы'])
MONTHS = morph_pipeline(['месяц','месяцев','месяца'])
EXPERIENCE = rule(
    NAME,
    NAME1,
    DASH.optional(),
    INT,
    YEARS,
    INT,
    MONTHS,
    DOT.optional()
    ).interpretation(
    Experience
)
```

```python
show_matches(EXPERIENCE, 'Женщина, 29 лет, родилась 19 ноября 1986, 65 000\nруб, Опыт работы — 2 года 1 месяц')
```
